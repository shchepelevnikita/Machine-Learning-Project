```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

# Data Load and Preparation

```python
#импортируем подготовленный датасет
pokemon_df = pd.read_csv('../data/pokemon_preprocessed.csv')
```

```python
#убеждаемся, что загрузка прошла успешно
pokemon_df.head()
```

```python
#нормируем вещественные признаки
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(pokemon_df.iloc[:,0:7])
scaled_df = pd.DataFrame(np.array(scaler.transform(pokemon_df.iloc[:,0:7])), columns = ["Total","HP","Attack","Defense","Sp_Atk","Sp_Def","Speed"], index = pokemon_df.index)
```

```python
pokemon_df = pd.concat([scaled_df, pokemon_df.iloc[:,7:]], axis = "columns")
pokemon_df
print(len(pokemon_df.columns))
```

```python
#создадим датафреймы X и y
y = pokemon_df.Legendary
pokemon_df = pokemon_df.drop(['Legendary'], axis=1)
X = pokemon_df
```

```python
from sklearn.model_selection import train_test_split
#разбиваем всю выборку на обучающую и тестовую
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state = 2)
```

# Finding suitable architecture

## Architecture 1

```python
clf = Sequential()
clf.add(Flatten())
clf.add(Dense(20, activation='relu', input_dim=len(X.columns)))
clf.add(Dense(30, activation='relu'))
clf.add(Dense(30, activation='relu'))
clf.add(Flatten())
clf.add(Dense(1, activation='softmax'))
```

```python
clf.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
print('\n')
history = clf.fit(X_train,y_train, batch_size=10, epochs=100)
print('\n')
print(clf.summary())
```

```python
print('Evaluating the classifier...')
eval_model = clf.evaluate(X_test, y_test)
print('Accuracy: ', eval_model[1])
print('Loss: ', eval_model[0])
print('\n ')
```

## Architecture 2

```python
 clf = Sequential()
    
#First Hidden Layer
clf.add(Dense(4, activation='relu', input_dim=len(X.columns)))
    
#Second  Hidden Layer
clf.add(Dense(4, activation='relu'))   
    
    #Third  Hidden Layer
clf.add(Dense(4, activation='relu'))

    #Output Layer
clf.add(Dense(1, activation='sigmoid'))
    
    #Compiling the neural network
clf.compile(optimizer ='adam',loss='binary_crossentropy', metrics =['accuracy'])
        # As this is a binary classification problem, we use binary_crossentropy 
        # to calculate the loss function between the actual output and the predicted output.  
        # As optimizer we use Adaptive moment estimation (ADAM)
print('\n')
history = clf.fit(X_train,y_train, batch_size=10, epochs=100)
print('\n')
print(clf.summary())
```

```python
print('Evaluating the classifier...')
eval_model = clf.evaluate(X_test, y_test)
print('Accuracy: ', eval_model[1])
print('Loss: ', eval_model[0])
print('\n ')
```

## Architecture 3

```python
clf = Sequential()

clf.add(Dense(128, activation='relu'))
clf.add(Dense(256, activation='relu'))
clf.add(Dense(1, activation='sigmoid'))

clf.compile(optimizer = 'adam',
             loss = 'binary_crossentropy',
             metrics = ['accuracy']
)
print('\n')
history = clf.fit(X_train,y_train, batch_size=10, epochs=100)
print('\n')
print(clf.summary())
print('Evaluating the classifier...')
eval_model = clf.evaluate(X_test, y_test)
print('Accuracy: ', eval_model[1])
print('Loss: ', eval_model[0])
print('\n ')
```

# Tuning hyperparameters

```python
from kerastuner.tuners import RandomSearch
```

```python
def build_model(hp):
    model = Sequential()
    model.add(Dense(units=hp.Int('units',
                                        min_value=64,
                                        max_value=192,
                                        step=64),
                           activation='relu'))
    model.add(Dense(units=hp.Int('units',
                                        min_value=128,
                                        max_value=384,
                                        step=128),
                           activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy'])
    return model
```

```python
tuner = RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=3,
    executions_per_trial=3,
    overwrite=True)
```

```python
tuner.search(X_train, y_train, epochs=40, validation_split=0.2)
```

```python
best_hps=tuner.get_best_hyperparameters(num_trials=1)[0]
```

```python
model = tuner.hypermodel.build(best_hps)
history = model.fit(X_train, y_train, epochs=40, validation_split=0.2)
```

```python
eval_result = model.evaluate(X_test, y_test)
print("[test loss, test accuracy]:", eval_result)
```

# Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix

clf = GradientBoostingClassifier()
clf.fit(X_train, y_train)
```
