```python
#импортируем библиотеку pandas
import pandas as pd
```

```python
#загружаем датасет
twitch_df = pd.read_csv('../data/twitch_data.csv')
```

```python
#убеждаемся, что все правильно загрузили и заодно смотрим на часть данных в датасете
twitch_df.head()
```

```python
#смотрим информацию о датасете
twitch_df.info()
```

Пропущенных данных в датасете нет. Поэтому приступим к рассмотрению других признаков, кроме вещественных. Кратко говоря, задача сформулирована так, что необходимо предсказать вещественное значение по другим вещественным значениям. Поэтому необходимо избавиться от всех признаков, не имеющих вещественный тип. Задача была так сформулирована, потому что был проведен анализ по данным и было выяснено, что такие признаки, как "Partnered", "Mature", "Language", "Channel" не влияют или почти не влияют на конечный результат. "Channel", очевидно, является неинформативным признаком. 

```python
#удаляем столбец Channel
twitch_df = twitch_df.drop(['Channel'], axis=1)
```

В следующей ячейке видно, что признак "Partnered" является незначительным, поэтому его тоже можно удалить, ведь из 1000 стримеров лишь 22 не имеют партнерства с Twitch.

```python
print(twitch_df[twitch_df['Partnered'] == 1].sum())
twitch_df = twitch_df.drop(['Partnered'], axis=1)
```

```python
print(list(twitch_df[twitch_df['Mature'] == 1].index))
```

По данным сверху видно, что стримеры, имеющие метку 18+, распределены более-менее равномерно по всему топу, что означает, что данный признак также неинформативен.

```python
twitch_df = twitch_df.drop(['Mature'], axis=1)
```

```python
for lang in twitch_df.Language.unique():
    temp = list(twitch_df[twitch_df['Language'] == lang].index)
    print(f'{lang} : {temp}')
```

Из этих данных так же можно понять, что язык не является особо информативным признаком, у популярных языков стримеры, говорящие на них, распределены более-менее равномерно по всему топу, а стримеров, ведущих трансляции на менее популярных языках, слишком мало, данные по ним незначительны. Поэтому данный признак тоже удалим.

```python
twitch_df = twitch_df.drop(['Language'], axis=1)
```
