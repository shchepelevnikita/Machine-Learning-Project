```python
#импортируем библиотеку pandas
import pandas as pd
```

```python
#загружаем датасет
twitch_df = pd.read_csv('../data/twitch_data.csv')
```

```python
#убеждаемся, что все правильно загрузили и заодно смотрим на часть данных в датасете
twitch_df.head()
```

```python
#смотрим информацию о датасете
twitch_df.info()
```

Пропущенных данных в датасете нет. Поэтому приступим к кодированию категориальных признаков. У нас их два: никнейм стримера ('Channel') и язык, на котором ведется трансляция ('Language'). Первый признак не является информативным, поэтому его можно удалить. Второй признак закодируем с помощью one-hot кодирования.

```python
#удаляем столбец Channel
twitch_df = twitch_df.drop(['Channel'], axis=1)
```

```python
#one-hot кодирование
twitch_df = pd.get_dummies(twitch_df, prefix=['lang'], columns=['Language'])
```

```python
#проверяем правильность датафрейма
twitch_df.head()
```

One-Hot кодирование было выполнено правильно. Порядковых признаков нет, поэтому приступим к кодированию бинарных. В датафрейме имеется два бинарных признака: "Partnered" и "Mature". Первый показывает то, имеет ли стример партнерство с платформой Twitch, второй же показывает то, имеет ли стример метку 18+.

```python
twitch_df['Partnered'] = twitch_df['Partnered'].astype('int64', copy=False)
twitch_df['Mature'] = twitch_df['Mature'].astype('int64', copy=False)
```

```python
#экспортируем изменения в датафрейме
twitch_df.to_csv('../data/twitch_data_preprocessed.csv', index=False)
```
